```python
import pandas as pd 
import matplotlib.pyplot as plt 
from collections import Counter
```

```python
data = pd.read_csv("../data/prepered_data/prepered_data.csv")
data.head()
```

```python
counter = Counter()
data['ner-tags'].apply(lambda x : eval(x)).apply(lambda x : counter.update(x))
print()
```

```python
counter
```

```python
print(counter.get('O'))
counter.__delitem__("O")
```

```python
[*counter.items()]
```

```python
sorted_count = sorted(counter.items() , key = lambda x : x[1] , reverse=True)
sorted_count = dict(sorted_count)
```

```python
plt.bar(x = sorted_count.keys() , height=sorted_count.values())
plt.xticks(rotation = 90)
plt.xlabel('Entities', color = 'red')
plt.ylabel('Count', color = 'red')
plt.show()
```
